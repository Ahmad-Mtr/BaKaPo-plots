```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_excel('data/employees_jordan_data.xlsx')
# 1. Line Chart - Average Salary Trend by Joining Date
plt.figure(figsize=(12, 6))
data.sort_values('Experience (Years)').plot(x='Experience (Years)', y='Salary', kind='line', marker='o')
plt.title('Salary Trend by Joining Date')
plt.xlabel('Joining Date')
plt.ylabel('Salary')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
